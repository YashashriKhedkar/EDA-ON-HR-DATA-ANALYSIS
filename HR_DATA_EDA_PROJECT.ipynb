###Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

##Chart 1

##Count the occurrences of each marital status

```python
# count MaritalStatus
MaritalStatus_count = df['MaritalStatus'].value_counts()
MaritalStatus_count
```

```python
# piechart is used for visualization
MaritalStatus_count.plot.pie(explode=[0.03, 0.03, 0.03], autopct='%1.2f%%', shadow = True, figsize=(8,5), fontsize=12, colors = ['#40E0D0', '#6495ED', '#DFFF00'])
plt.title('Marital Status Count',fontsize = 15)
plt.show()
```

##Chart 2

##Monthly Salary Comparison Across Job Roles by Gender


```python
# Chart - 2 visualization code
job_role_count = df.groupby(['JobRole', 'Gender'])['MonthlySalary'].mean().reset_index()
job_role_count['MonthlySalary'] = job_role_count['MonthlySalary'].astype(int)
job_role_count
```

```python
# barplot is used for visualization
plt.figure(figsize=(15,8))
ax = sns.barplot(data=job_role_count, x='MonthlySalary', y='JobRole', hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
plt.title('Gender vise JobRole with MonthlySalary', fontsize=15)
plt.show()
```

##Chart 3

## Department-wise Monthly Salary Distribution by Attrition

```python
department_attrition = df.groupby(['Department','Attrition'])['MonthlyIncome'].mean().reset_index()
department_attrition['MonthlyIncome'] = department_attrition['MonthlyIncome'].astype(int)
department_attrition
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(data=department_attrition, x='Department', y='MonthlyIncome', hue='Attrition')
plt.title('Attrition vise Department VS MonthlyIncome', fontsize=15)
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

##Chart 4

##Average Employee Count by Education Field

```python
count_hourlysal = df.groupby(['EducationField'])['EmployeeNumber'].mean().reset_index().sort_values(by='EmployeeNumber', ascending=False)
count_hourlysal
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(x=df['EducationField'], y=df['EmployeeNumber'])
plt.title("Education Field vise employee count", fontsize=18)
plt.xlabel("EducationField", fontsize=15)
plt.ylabel("Employee Count", fontsize=15)
plt.show()
```

##Chart 5

##Daily Salary by Age Group

```python
def categories_age(Age):
  if Age <= 25:
    return '18-25'
  elif Age > 25 and Age <= 35:
        return '26-35'
  elif Age > 35 and Age <= 45:
        return '36-45'
  elif Age > 45 and Age < 55:
        return '46-55'
  else:
        return '55+'

df['AgeGroup'] = df['Age'].apply(categories_age)
df.head(2)
```

```python
Salary_AgeGroup = df.groupby(['AgeGroup'])['DailySalary'].mean().reset_index().sort_values(by='DailySalary', ascending=False)
Salary_AgeGroup['DailySalary'] = Salary_AgeGroup['DailySalary'].astype(int)
Salary_AgeGroup
```

```python
plt.figure(figsize=(10,4))
ax = sns.barplot(x=Salary_AgeGroup['AgeGroup'],y=Salary_AgeGroup['DailySalary'], color='pink')
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```
